## Part 0: Executive Summary

### The Story So Far

Update 1 closed on April 30 with a clean thesis: physical oil decoupled from financial assets during the 33-day Hormuz closure. WTI rose +32.9% while energy equities, gold, and airlines diverged or fell. The mechanism was physical scarcity — the strait carried only ~2% of normal transit volume.

### What Update 2 Adds

**The May Paradox.** WTI fell nearly 20% in May — its worst month since COVID. The strait remained at ~2% capacity throughout. Physical supply conditions were unchanged. What changed was the forward expectation of peace. Ceasefire diplomacy overrode scarcity pricing. Markets started pricing the barrel that *would* move through an eventually re-opened strait rather than the barrel that *couldn't* move today.

**The June 10 Re-Escalation.** Iran announced a full commercial closure on June 10. WTI spiked on Trump's ceasefire rejection, then fell on renewed talks. This is the second testable shock event — same framework, measurable whether the market had learned a playbook.

**Three new analytical stories run in parallel:**

## Part 2: What Changed Since April 30 — The Full Timeline

Update 1 ended with WTI at its peak and a clean supply-shock regime. Three things happened afterward that fundamentally change the analytical picture:

**1. The May Paradox (May 1–29):** Strait at 2% capacity. WTI fell ~20% anyway. Diplomacy pricing overrode physical scarcity — the single most surprising finding of the entire 2026 study period.

**2. June 10 Re-escalation:** Iran's second closure announcement. WTI spiked, then fell on renewed ceasefire optimism. This is now the *sixth testable event* the framework can analyze.

**3. New asset signals:** ITA revealed the backlog dynamic. BWET showed a freight premium entirely above and beyond oil prices. TLT tested whether bonds complete the safe-haven failure trifecta.

## Part 5: The Paradox Correction — When Supply Stopped Mattering

**Setup:** Strait still at ~2% of normal transit capacity throughout May. No new barrels started moving. The physical disruption that drove WTI to +32.9% in March–April was structurally unchanged.

**What happened:** WTI dropped ~20% in May. Forward markets were pricing the barrel that *would* eventually flow through a re-opened strait, not the barrel that couldn't move today. This is the regime shift — from **scarcity pricing** to **expectations pricing**.

**The WTI/Gold ratio is the cleanest signal.** It peaked on April 7 (first ceasefire announcement) and declined through May even as the strait remained closed. That peak is the exit signal for the Long WTI / Short XOM pair trade — not the official ceasefire, which came later with more noise.

| WTI/Gold Rising | WTI/Gold Falling |
|-----------------|-----------------|
| Supply shock dominant | Diplomacy dominant |
| Oil outpaces gold | Gold holds / Oil corrects |
| Physical scarcity pricing | Expectations pricing |
| Long WTI pair: entry signal | Long WTI pair: exit signal |

## Part 6: June 10 Re-Escalation — Second Shock Test

**The question:** By June 10, the market had been living with the Hormuz crisis for 102 days. Did the second escalation trigger the same magnitude response as February 28?

**Market learning hypothesis:** If the June 10 shock is smaller in magnitude or faster to reprice, it suggests the market had internalised a playbook. Oil up → equities down → defense up. If similar magnitude, no learning occurred — each escalation is treated as fresh information.

**Escalation Replay** shows 5-day cumulative return arcs for both shock events side-by-side across key assets. The pattern comparison is the finding.

## Appendix A: Methodology

### A.1 Framework

Same OLS market model as Update 1, extended to 5 windows.

$$AR_{i,t} = R_{i,t} - (\hat{\alpha}_i + \hat{\beta}_i \cdot R_{m,t}) \qquad CAR_i = \sum AR_{i,t}$$

**Estimation window:** Jan 1–Feb 27, 2026 (39 trading days). Beta estimated here, applied to all four post-event windows. Caveat: escalation risk was already being priced during the estimation window — betas may understate pre-shock energy sensitivity.

**WTI/Gold regime signal:**
$$\text{Ratio}_t = \frac{P_{WTI,t}}{P_{GOLD,t}} \text{ (base 100 at Jan 1, 2026)}$$

Rising → supply shock. Falling → diplomacy/expectations.

### A.2 Window Boundaries

| Window | Dates | Driver |
|--------|-------|--------|
| Pre-event | Jan 1–Feb 27 | Escalation pricing |
| Shock | Feb 28–Apr 16 | Physical scarcity |
| False Dawn | Apr 17–Apr 30 | Peace signal |
| Paradox Correction | May 1–May 29 | Diplomacy beats supply |
| Volatile Diplomacy | Jun 1–Jun 13 | Re-escalation + talks |

## Appendix B: Limitations That Matter
